```python
train_data_file = "./tranfered_data/train.csv"
test_data_file = "./tranfered_data/test.csv"
df_tr = pd.read_csv(train_data_file)
df_ts = pd.read_csv(test_data_file)
################################ processing null #############
#for col in df_tr.keys():
#    df_tr[col] = df_tr[col].fillna(-99, inplace=True)
#for col in df_ts.keys():
#    df_ts[col] = df_ts[col].fillna(-99, inplace=True)
```

```python
redun_tag = [x for x in df_tr.keys() if "bad_flag_worst6" in x]# 'userid', "PosNgTag'
```

```python
print ['userid', 'PosNgTag'] + redun_tag
```

```python
df_mx_tr = df_tr.drop(['userid', 'PosNgTag'] + redun_tag, 1)
df_my_tr = df_tr[['PosNgTag']]


mx_tr = df_mx_tr.values
my_tr = df_my_tr.values

df_mx_ts = df_ts.drop(['userid', 'PosNgTag'] + redun_tag, 1)
df_my_ts = df_ts[['PosNgTag']]

mx_ts = df_mx_ts.values
my_ts = df_my_ts.values
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification
```

```python
model = xgboost.XGBClassifier()
#model = RandomForestClassifier(max_depth=6, random_state=0)
model.fit(mx_tr, my_tr)
print(model)
y_pred = model.predict(mx_ts)
predictions = [round(value) for value in y_pred]
accuracy = accuracy_score(my_ts, predictions)
print("Accuracy: %.2f%%" % (accuracy * 100.0))
```
